```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings

%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 12
warnings.filterwarnings('ignore')
```

### Merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df1 = pd.read_csv('college_1.csv')
df2 = pd.read_csv('college_2.csv')

# Merge the dataframes to one
df = pd.concat([df2, df1], axis=0)

df
```

```python
# Replace nan with 0
df = df.replace(np.nan,0)
df
```

#### 1. Consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


```python
exceeded_expectations = df.loc[df['CodeKata Score'] >= 15000]

exceeded_expectations
```

```python
exceeded_expectations.to_csv('Exceeded_Expectations.csv')
print('A new csv file is generated')
```

#### 2. If  10000<codekata score<15000   (Reached_expectations.csv)



```python
reached_expectations = df.loc[(df['CodeKata Score']> 10000) & (df['CodeKata Score'] < 15000)]
reached_expectations
```

```python
reached_expectations.to_csv('Reached_Expectations.csv')
print('A new csv file has been generated .../')
```

#### 3. If  7000<codekata score<10000   (Needs_Improvement.csv)

#### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
df.Department.unique()
```

```python
EE = df[df['Department']=='Electronics and Electrical Engineering'].sort_values('CodeKata Score', ascending=False).head(3)
CE = df[df['Department']=='Computer Science and Engineering'].sort_values('CodeKata Score', ascending=False).head(3)
EC = df[df['Department']=='Electronics and Communication Engineering'].sort_values('CodeKata Score', ascending=False).head(3)

highest = pd.concat([EE, CE, EC], axis=0)
highest
```

```python
# bar chart of department wise toppers 

fig = px.bar(
    highest,
    x="Department",
    y=["Name", "CodeKata Score"],
    color='Name',
    barmode='group',
    title='Department wise Toppers',
    
)

fig.update_layout(
    height=700,
    width=1300,
    font_size=14,
    font_family ='Monospace',
    legend_title_font_color='blue',
    title= dict(
        x=0.1,
        y=0.95,
        font= {
            "family":'Time New Roman',
            "size":30,
            "color":'black'
        }
    ),
)
fig.show()
```
